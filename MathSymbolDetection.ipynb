## Faster RCNN with ResNet-50 FPN

```python
from detecto import core, utils, visualize
from detecto.visualize import show_labeled_image, plot_prediction_grid
from torchvision import transforms
import matplotlib.pyplot as plt
import numpy as np
```

```python
custom_transforms = transforms.Compose([
transforms.ToPILImage(),
transforms.ColorJitter(brightness=0.2, contrast=0.2, saturation=0.2, hue=0.2)
transforms.ToTensor(),
utils.normalize_transform(),
])
```

```python
Train_dataset=core.Dataset('/content/drive/MyDrive/Annotated/train_xml',transform=custom_transforms)
Test_dataset = core.Dataset('/content/drive/MyDrive/Annotated/valid_xml')
loader=core.DataLoader(Train_dataset, batch_size=2, shuffle=True)
model = core.Model(['math'])
losses = model.fit(loader, Test_dataset, epochs=30, lr_step_size=5, learning_rate=0.001, verbose=True)
```

```python
plt.plot(losses)
plt.show()
```

```python
print(losses)
```

```python
model.save('/content/drive/MyDrive/Models/fasterrcnn_resnet50fpn.pth')
# model = core.Model.load('/content/drive/MyDrive/Models/fasterrcnn1.pth', ['math'])
```

```python
image = utils.read_image('/content/drive/MyDrive/TC11_package/CROHME2011/test/Inkdata_temp_InkFR_HPR_EQU_NOC_scc784_fi4_db137833.png') 
predictions = model.predict(image)
labels, boxes, scores = predictions
show_labeled_image(image, boxes, labels)
```

```python
thresh=0.9
filtered_indices=np.where(scores>thresh)
filtered_scores=scores[filtered_indices]
filtered_boxes=boxes[filtered_indices]
num_list = filtered_indices[0].tolist()
filtered_labels = [labels[i] for i in num_list]
show_labeled_image(image, filtered_boxes, filtered_labels)
```

```python
import cv2
import torch
import matplotlib.pyplot as plt
import matplotlib.patches as patches
from detecto import core
import numpy as np
```

```python
from google.colab.patches import cv2_imshow
```

```python
model = core.Model.load('/content/drive/MyDrive/Models/fasterrcnn_resnet50fpn.pth', ['math'])
```

```python
image = cv2.imread('/content/drive/MyDrive/TC11_package/CROHME2011/test/Inkdata_temp_InkFR_HPR_EQU_NOC_scc77_fi5_db138964.png') 
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
predictions = model.predict(image)
labels, boxes, scores = predictions

thresh=0.9
filtered_indices=np.where(scores>thresh)
filtered_scores=scores[filtered_indices]
filtered_boxes=boxes[filtered_indices]
num_list = filtered_indices[0].tolist()
filtered_labels = [labels[i] for i in num_list]
```

```python
cv2_imshow(image)
```

```python
rectlist = []
for box in filtered_boxes:
  x, y = int(box[0]), int(box[1])
  endx, endy = int(box[2]), int(box[3])
  cv2.rectangle(image, (x,y), (endx, endy), (0,255,0), 2)
  rectlist.append((x,y,endx,endy))
```
